```python
from sympy.interactive import printing
printing.init_printing(use_latex=True)
import sympy as sym
```

```python
eqs = [Eq(x**2 + 3*y, 10),
      Eq(4*x - y**2, -2)]
solve(eqs)
```

## Zadanie 3

```python
s = solve(eqs)
print("Ukad ma", len(s), "rozwizania")
```

## Zadanie 4

```python
for i in s:
    for k, v in i.items():
        print(k, v.evalf())
    print("\n")
```
